```python
import matplotlib.pyplot as p
import numpy as np
x = np.arange(0, 5*np.pi, 0.1)
y = np.cos(x)
print(p.plot(x,y))
```

```python
x = np.arange(0, 5*np.pi, 0.1)
y = np.tan(x)
print(p.plot(x,y))
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])
print(p.plot(ypoints, linestyle='dashed'))
p.title('Sample graph')
p.xlabel('X axis')
p.ylabel('Y axis')
print(p.show())
```
