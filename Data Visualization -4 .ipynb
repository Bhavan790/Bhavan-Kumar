```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
X = np.array([1, 2, 6, 8, 12, 14, 20])
Y = np.array([3, 8, 1, 10, 12, 16, 18])
Z = np.random.rand(len(X)) 
fig = plt.figure(figsize=(12, 8))
ax = plt.axes(projection='3d')
p = ax.scatter3D(X, Y, Z)
plt.colorbar(p)
plt.show()

# 2D Scatter Plot
X2 = np.array([10, 80, 100, 140, 180, 190, 280])
fig2 = px.scatter(x=X2, y=np.random.rand(len(X2)))  # Adding random Y values for 2D plotting
fig2.show()
```
